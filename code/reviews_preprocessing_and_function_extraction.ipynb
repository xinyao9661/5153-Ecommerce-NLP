```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('data/reviews_full_cond and sham_no duplicates.csv', index_col=0)
data = data.rename(columns={'Unnamed: 0.1':'asin'}).reset_index(drop=True)
data.head()
```

```python
import math
math.isnan(data.iloc[0,2])
```

```python
asin = []
reviews = []
for i in range(len(data)):
    for j in range(1,1350):
        try:
            if math.isnan(data.iloc[i,j])==True:
                break
        except TypeError:
            asin.append(data.asin[i])
            reviews.append(data.iloc[i,j])
```

```python
df = pd.DataFrame({'asin': asin, 'raw_info': reviews})
df
```

```python
import ast
df.raw_info = df.raw_info.apply(ast.literal_eval)
df
```

```python
df.raw_info[1]
```

```python
def helpful(x):
    return x['helpful']

def review(x):
    return x['reviewText']

def overall(x):
    return x['overall']

def summary(x):
    return x['summary']

def unixReviewTime(x):
    return x['unixReviewTime']

def reviewTime(x):
    return x['reviewTime']

def reviewerID(x):
    return x['reviewerID']
```

```python
df['review'] = df.raw_info.apply(review)
df['overall'] = df.raw_info.apply(overall)
df['helpful'] = df.raw_info.apply(helpful)
df['summary'] = df.raw_info.apply(summary)
df['unixReviewTime'] = df.raw_info.apply(unixReviewTime)
df['reviewTime'] = df.raw_info.apply(reviewTime)
df['reviewerID'] = df.raw_info.apply(reviewerID)
```

```python
df.drop('raw_info', axis='columns', inplace=True)
df
```

```python
import string
def remove_punctuation(text):
    text = ''.join([c for c in text if c not in string.punctuation])
    return text.lower()
```

```python
data['review'] = data['review'].apply(remove_punctuation)
data['title'] = data['title'].apply(remove_punctuation)
data['summary'] = data['summary'].apply(remove_punctuation)
data['helpful_overall'] = data.helpful.apply(lambda x: x[0])
data['helpful_positive'] = data.helpful.apply(lambda x: x[1])
```

```python
# categorize functions
func_cleansing = ['purifying', 'clarifying', 'cleansing', 'cleanse', 'cleanser', 'clean', 'detox']
func_dandruff = ['dandruff', 'antidandruff', 'scalp', 'psoriasis']
func_growth = ['growth', 'grow', 'thickening', 'thicken', 'thick', 'thicker', 'volumizing', 'volume', 'loss', 
               'amplify', 'amplifying', 'extension','extensions']
func_moisture = ['moisture', 'moisturizing', 'hydrating', 'hydration', 'hydrate', 'moist'
                 'smooth', 'smoothen', 'smoothing', 'nourish', 'nourishing']
func_repair = ['strength', 'strengthen', 'strengthening','antibreakage', 'repair', 'restore', 'damage', 'treatment', 'treatments', 
               'recovery', 'recover', 'fortify', 'fortifying', 'restructure', 'restructuring', 'revitalize', 'revitalizing',
              'enhancing', 'protective']
func_color = ['color', 'colour', 'color treated', 'colour treated']
functions = [func_cleansing, func_dandruff, func_growth, func_moisture, func_repair, func_color]
```

```python
title_funcs = []
review_funcs = []

for j in range(len(functions)):
    title_func = []
    review_func = []
    for i in range(len(data)):
        t = data.title[i]
        r = data.review[i]
        func_list = functions[j]
        t_total = []
        r_total = []
        for w in func_list:
            if w in list(t.split()):
                t_total.append(w)
            if w in list(r.split()):
                r_total.append(w)
            else:
                pass
        if len(t_total)>0:
            title_func.append(1)
        else:
            title_func.append(0)
        if len(r_total)>0:
            review_func.append(1)
        else:
            review_func.append(0)
    title_funcs.append(title_func)
    review_funcs.append(review_func)
```

```python
len(title_funcs[0])
```

```python
func_names = ['func_cleansing', 'func_dandruff', 'func_growth', 'func_moisture', 'func_repair', 'func_color']

# new_columns
for i in range(len(func_names)):
    data['t_'+func_names[i]] = title_funcs[i]
    data['r_'+func_names[i]] = review_funcs[i]
```

```python
for j in range(len(func_names)):
    satisfactory = []
    for i in range(len(data)):
        lst = []
        if data['t_'+func_names[j]][i] == data['r_'+func_names[j]][i]:
            lst.append(1)
        else:
            lst.append(0)
        if len(lst) == sum(lst):
            satisfactory.append(1)
        else:
            satisfactory.append(0)
len(satisfactory)
```

```python
# if a review mentions all the functions in product name, then 1, else 0
data['satisfactory'] = satisfactory
```

```python
data['temp'] = sum([data['t_'+func_names[i]] for i in range(len(func_names))])
```
