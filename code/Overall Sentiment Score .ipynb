```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('meta_and_review.csv')
```

```python
df.drop(['Unnamed: 0', ''], axis=1, inplace=True)
```

```python
text_df = pd.DataFrame (columns= ['index', 'review'])
text_df.loc[:,'index'] = df.index
text_df.loc[:,'review'] = df.review
```

```python
text_df.head()
```

```python
# Load libraries
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data
SIA = SentimentIntensityAnalyzer()
text_df['polarity_score']=text_df.review.apply(lambda x:SIA.polarity_scores(str(x))['compound'])
```

```python
text_df['neutral_score']=text_df.review.apply(lambda x:SIA.polarity_scores(str(x))['neu'])
text_df['negative_score']= text_df.review.apply(lambda x:SIA.polarity_scores(str(x))['neg'])
text_df['positive_score']= text_df.review.apply(lambda x:SIA.polarity_scores(str(x))['pos'])
#text_df['sentiment']=''
#text_df.loc[text_df.polarity_score>0,'sentiment']='POSITIVE'
#text_df.loc[text_df.polarity_score==0,'sentiment']='NEUTRAL'
#text_df.loc[text_df.polarity_score<0,'sentiment']='NEGATIVE'
text_df.head()
```

```python
df = df.merge(text_df, left_index=True, right_index=True).rename(columns= {'review_x':'review'
                                                                     })
df.drop(['review_y', 'index'],axis=1,inplace=True)
#polarity: overall
#neural
#pos
#neg
```
